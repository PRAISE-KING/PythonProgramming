```python
# grouping the whole data sets into features (X) and outcome(Y)

array = data.values

# Features
X = array[0:769, 0:8] # IF YOU WANT ALL DATA for column then [:, 0:4] or for all rows and columns [: , :] can also do for rows alone 
X
```

```python
# from IPython.utils.tempdir import NamedFileInTemporaryDirectory
# Using sklearn library

from sklearn import model_selection #model_selection used to split data into training sets(70%) and testing sets(30%)

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=40)
# our training is 70% (0.70)
# testing sets is 30% (0.30)

# X and Y have been subdevided into four portions Namely:
# X_train, X_test, Y_train, Y_test
```

```python
# Training phase
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

# cross validation

```

```python
model = RandomForestClassifier()
# Modet Training - training features(X_train) with their correct outcome (Y_train)

model.fit(X_train, Y_train)

```

```python
# model evaluation/testing

predictions = model.predict(X_test)
print("the model predicted",predictions)
```

```python
# correct answers

print("The correct answers :", Y_test)
```

```python
# Scientific Model Testing
# Accuracy_score
# Confusion Matrix

from sklearn.metrics import accuracy_score

print("The model accuracy is : ", accuracy_score(predictions, Y_test)*100)
```
